__4.__ Решить систему линейных уравнений методом Холецкого

-45x_{1}+50x_{2}-15x_{3}=-460 \\

\end{cases}$$

$A =\begin{pmatrix}

$l_{31} = 1/l_{11} \cdot a_{31} = 1/9 \cdot (45) = 5 $ <br>
$l_{22} = \sqrt{a_{22} - l_{21}^2} = \sqrt{50 - 25} = 5 $ <br>
$l_{32} = 1/l_{22} \cdot (a_{32} - l_{21}l_{31}) = 1/5 \cdot(-15 -(-5)\cdot 5) = 2$ <br>
$l_{33} = \sqrt{a_{33} - l_{31}^2 - l_{32}^2} = \sqrt{38 - 25 - 4} = 3$ <br>

$L = \begin{pmatrix}
9 & 0 & 0 \\
-5 & 5 & 0 \\
5 & 2 & 3
\end{pmatrix}$, 
$L^{T} = \begin{pmatrix}
